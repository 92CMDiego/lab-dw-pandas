### Data Description

- Customer - Customer ID

- ST - State where customers live

- Gender - Gender of the customer

- Education - Background education of customers 

- Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer. In other words, is the predicted or calculated value of a customer over their entire duration as a policyholder with the insurance company. It is an estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company. Customer Lifetime Value takes into account factors such as the duration of the customer's policy, premium payments, claim history, renewal likelihood, and potential additional services or products the customer may purchase. It helps insurers assess the long-term profitability and value associated with retaining a particular customer.

- Income - Customers income

- Monthly Premium Auto - Amount of money the customer pays on a monthly basis as a premium for their auto insurance coverage. It represents the recurring cost that the insured person must pay to maintain their insurance policy and receive coverage for potential damages, accidents, or other covered events related to their vehicle.

- Number of Open Complaints - Number of complaints the customer opened

- Policy Type - There are three type of policies in car insurance (Corporate Auto, Personal Auto, and Special Auto)

- Vehicle Class - Type of vehicle classes that customers have Two-Door Car, Four-Door Car SUV, Luxury SUV, Sports Car, and Luxury Car

- Total Claim Amount - the sum of all claims made by the customer. It represents the total monetary value of all approved claims for incidents such as accidents, theft, vandalism, or other covered events.

```python
# Your code here

import pandas as pd
import numpy as np

df0 = pd.read_csv('customers.csv')

df0.shape

df0.dtypes
```

```python
# We can see that customer lifetime value and number of open complaints appear as objects. They should be numbers. We need to address that.

df0
```

```python
# First we need to clean the data.
# We see that there a number of rows (2937) that contain no information (neither in column 'Customer' nor in the rest of the columns).
# We delete them.

df0 = df0[df0['Customer'].isna()==False]
df0.isna().sum()
df0

# Now we need to have unique values in the columns where is needed and to get rid of the NaN values and probably of some of the 0 values.
# So we keep going with the cleaning.
```

```python
# We remove the "%" symbol from 'Customer Lifetime Value' column in order to be able to change it to numeric.

df0['Customer Lifetime Value'] = df0['Customer Lifetime Value'].str.replace('%', '')
df0['Customer Lifetime Value'].isna().sum()
df0.dtypes
```

```python
# Now we're able to change 'Customer Lifetime Value' column to numeric.

df0['Customer Lifetime Value'] =  pd.to_numeric(df0['Customer Lifetime Value'], errors='coerce')
df0.dtypes
```

```python
# We put all column names in lower case.

df0.columns = [col.lower() for col in df0.columns]
```

```python
# we use bfill method to fill na values of gender column.

df0['gender']=df0['gender'].fillna(method='bfill')
```

```python
df0['gender'].unique()
```

```python
# Now we put gender values in uppercase and transform them in order to have unique values in that column.

df0['gender'] = list(map(str.upper, df0['gender']))

def clean(x):
    if x in ['M', 'MALE']:
        return 'Male'
    elif x.startswith('F'):
        return 'Female'
    else:
        return 'U'

df0['gender'] = list(map(clean, df0['gender']))
```

```python
df0['gender'].unique()

df0['gender'].value_counts()

# Now we have unique values in gender column.
```

```python
# We split number of complaints column into three different columns. That way, we can see the values that are betwween slashs.

df0[['complaint_type1', 'complaint_type2', 'complaint_type3']] = df0['number of open complaints'].str.split('/', expand=True)
df0
```

```python
# And we change the datatype of these new columns to numeric.

df0['complaint_type1'] =  pd.to_numeric(df0['complaint_type1'], errors='coerce')
df0['complaint_type2'] =  pd.to_numeric(df0['complaint_type2'], errors='coerce')
df0['complaint_type3'] =  pd.to_numeric(df0['complaint_type3'], errors='coerce')

df0.dtypes
```

```python
# We change the names of 'st' column (WA = Washington, Cali = California, AZ = Arizona)

print(df0['st'].unique())

def clean_state(x):
    if x in ['WA']:
        return 'Washington'
    elif x in ['Cali']:
        return 'California'
    elif x in ['AZ']:
        return 'Arizona'
    else:
        return x
    
df0['st'] = list(map(clean_state, df0['st']))
```

```python
print(df0['st'].unique())

# Now we have unique values.
```

```python
# We do the same with 'education' column.

df0['education']=df0['education'].str.replace('Bachelors', 'Bachelor')
```

```python
print(df0['education'].unique())

# Now we have unique values.
```

```python
# Now we replace 0 values in column 'income' with the median of the values of that column 'income'.

income_median = df0['income'].median()
df0['income'] = df0['income'].replace(0, income_median)
```

```python
df0['income'].value_counts
```

```python
# For 'customer lifetime value', we choose to replace na values with the mean of the values of that column.

mean_cust_life = df0['customer lifetime value'].mean()
```

```python
df0['customer lifetime value'] = df0['customer lifetime value'].fillna(mean_cust_life)
```

```python
df0['customer lifetime value'].unique()
```

```python
df0

# Finally, we're done with the cleaning and we can start analyzing the data.
```

```python
print(df0.dtypes)
```

```python
# We calculate the range of the numerical columns.

range_df0_clv = df0['customer lifetime value'].max()-df0['customer lifetime value'].min()
print(range_df0_clv)

range_df0_income = df0['income'].max()-df0['income'].min()
print(range_df0_income)

range_df0_mpa = df0['monthly premium auto'].max()-df0['monthly premium auto'].min()
print(range_df0_mpa)

range_df0_tca = df0['total claim amount'].max()-df0['total claim amount'].min()
print(range_df0_tca)
```

```python
# We store numerical columns into a new dataframe an use describe to describe its statistics.

numeric_cols = df0._get_numeric_data()

numeric_cols

numeric_cols.describe()
```
